# Exploring Data from 2015 Axial Seamount Eruption

## Importing data packages:

```python
# Importing pandas, numpy, and pyplot

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Importing data sets from OOI:

```python
# Mean seafloor depth in meters (dep)

# Central Caldera (CC)
url1 = 'https://erddap.dataexplorer.oceanobservatories.org/erddap/tabledap/ooi-rs03ccal-mj03f-05-botpta301.csv?time%2Cbotsflu_meandepth%2Cz&time%3E%3D2014-08-29T20%3A59%3A00Z&time%3C%3D2019-12-31T00%3A00%3A00Z'

# Eastern Caldera (EC)
url2 = 'https://erddap.dataexplorer.oceanobservatories.org/erddap/tabledap/ooi-rs03ecal-mj03e-06-botpta302.csv?time%2Cbotsflu_meandepth%2Cz&time%3E%3D2014-09-03T22%3A05%3A00Z&time%3C%3D2019-12-31T23%3A59%3A00Z'
```

```python
# Mean seafloor pressure in psia (pounds per square inch absolute) (psia)

#Central Caldera (CC)
url3 = 'https://erddap.dataexplorer.oceanobservatories.org/erddap/tabledap/ooi-rs03ccal-mj03f-05-botpta301.csv?time%2Cbotsflu_meanpres%2Cz&time%3E%3D2014-08-29T20%3A59%3A00Z&time%3C%3D2023-03-14T08%3A30%3A00Z'

# Eastern Caldera (EC)
url4 = 'https://erddap.dataexplorer.oceanobservatories.org/erddap/tabledap/ooi-rs03ecal-mj03e-06-botpta302.csv?time%2Cbotsflu_meanpres%2Cz&time%3E%3D2014-09-03T22%3A05%3A00Z&time%3C%3D2023-03-14T08%3A31%3A00Z'
```

```python
# Reading csv file and turning them into dataframes

Cdep = pd.read_csv(url1, low_memory=False)
CCdep = pd.DataFrame(Cdep)

Edep = pd.read_csv(url2, low_memory=False)
ECdep = pd.DataFrame(Edep)

Cpsia = pd.read_csv(url3, low_memory=False)
CCpsia =pd.DataFrame(Cpsia)

Epsia = pd.read_csv(url4, low_memory=False)
ECpsia = pd.DataFrame(Epsia)
```

## Cleaning up data

```python
# Dropping the first row of all data sets because they contain units.
# Dropping the z column because it contains no meaningful info (all values are zero)

CCdep1 = CCdep.drop(index=0)
CCdep2 = CCdep1.drop(columns=['z'])

ECdep1 = ECdep.drop(index=0)
ECdep2 = ECdep1.drop(columns=['z'])

CCpsia1 = CCpsia.drop(index=0)
CCpsia2 = CCpsia1.drop(columns=['z'])

ECpsia1 = ECpsia.drop(index=0)
ECpsia2 = ECpsia1.drop(columns=['z'])
```

## Central Caldera Seafloor Depth

```python
# Converting to datetime and setting it as the index

CCdep2['time'] = pd.to_datetime(CCdep2["time"])
CCdep2.set_index('time', inplace = True)
CCdep2.head()
```

```python
# Coverting data type to float64

CCdep2['botsflu_meandepth'] = CCdep2['botsflu_meandepth'].astype('float64')
```

```python
# Making a plot

ax = CCdep2.loc['2015-04':'2015-05'].plot(figsize=(18,10))
ax.set_title('Central Caldera: Mean Seafloor Depth', fontsize=28)
ax.set_xlabel('Date', fontsize=22)
ax.set_ylabel('Depth below Sea Surface (m)', fontsize=22)
ax.legend(loc='upper right', fontsize=28, labels='meters')
```

Figure 1. Axial Seamount volcano erupted on April 24, 2015. The mean seafloor depth change of the Central Caldera from April 1st 2015 to May 1st 2015 is shown here. Depth is in meters below sea surface.
